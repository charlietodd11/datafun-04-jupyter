This week's project will use a dataset from the seaborn library to to practice working with exploritory data analysis. 

```python
#dependcies 
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

Data Acquisition and check data format

```python
#Load Iris dataset 
df = sns.load_dataset('iris')
print(df.head())
```

Initial Data Inspection

```python
print(df.head(10))
print('')

print(df.shape) 

print('')

print(df.dtypes)
```

Initial Descriptive Statistics

```python
print(df.describe())
```

Initial Data Distribution for Numerical Columns

```python
df['sepal_length'].hist()

df['sepal_width'].hist()

df['petal_length'].hist()

df['petal_width'].hist()

plt.legend(df)

plt.xlabel('Measurement (cm)')
plt.ylabel('Frequency')

plt.show()
```

Analysis:
In the histogram above, I have a few observations. First, each component of the iris has a clear distinction in size. The sepal length is the largest while the clear smallest is the petal width. The lenghts for the petal and sepal are both larger than their respective widths. It appears the petals have a greater spread than the sepals do. 

Initial Data Distribution for Categorical Columns

```python
#Inspect value counts by categorical column
df['species'].value_counts()
 
#Inspect value counts for all categorical columns
for col in df.select_dtypes(include = ['object', 'category']).columns:
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

plt.show()
```

Analysis: The Distibution of species indicates that all three species are euqally represented. 

Initial Data Transformation and Feature Engineering

```python
#Rename columns
df.rename(columns = {'sepal_length': 'Sepal Length'}, inplace = True)
df.rename(columns = {'sepal_width': 'Sepal Width'}, inplace = True)
df.rename(columns = {'petal_length': 'Petal Length'}, inplace = True)
df.rename(columns = {'petal_width': 'Petal Width'}, inplace = True)

#Adding new columns
df['Sepal Area'] = df['Sepal Length'] * df['Sepal Width']
df['Petal Area'] = df['Petal Length'] * df['Petal Length']
```

The following is a scatterplot of the area of the petals as compared to the sepal area for each species of iris

```python
plt = sns.scatterplot (data=df, x='Petal Area', y = 'Sepal Area', hue = "species")
plt.set_title("Petal Area vs Sepal Area")
```

Analysis: There exists a positive correlation between petal area and sepal area in the iris species versicolor and virginaica with a stronger correlation with veriscolor. The species setosa appears to have a small petal area regardless of sepal area.
